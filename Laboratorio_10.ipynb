### Construcción del modelo y Evaluación de Modelo

## Personas

Los tres modelos presentados utilizan técnicas de aprendizaje supervisado para abordar problemas de clasificación. El primer modelo emplea el clasificador de k vecinos más cercanos (KNN), donde la predicción para un nuevo punto se basa en la mayoría de las etiquetas de sus vecinos más cercanos en el espacio de características. Este enfoque es intuitivo y simple, adecuado para conjuntos de datos pequeños a medianos.

El segundo modelo implementa el clasificador de bosques aleatorios (Random Forest), que combina múltiples árboles de decisión para mejorar la generalización y reducir el sobreajuste. Al construir una variedad de árboles y promediar sus predicciones, el modelo proporciona robustez y eficacia, siendo particularmente efectivo para conjuntos de datos grandes y complejos.

El tercer modelo ajusta los hiperparámetros del clasificador de bosques aleatorios, incorporando una mayor cantidad de árboles, ajustando la profundidad máxima y los criterios de división. Además, utiliza validación cruzada para obtener una estimación más fiable del rendimiento del modelo y evalúa la importancia de las características en la predicción. Este enfoque más sofisticado busca optimizar la precisión y la capacidad de generalización del modelo en diversos contextos.

### Hogar
Se realizaron dos mejoras clave en el modelo. Primero, se agregó una capa BatchNormalization después de la capa Conv1D para normalizar los datos y estabilizar el entrenamiento. Segundo, se introdujo una capa Dropout antes de la capa Dense con una tasa del 50%, mejorando la capacidad del modelo para generalizar y reduciendo el riesgo de sobreajuste durante el entrenamiento.

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

model = models.Sequential()

model.add(layers.Reshape((127, 1), input_shape=(127,)))
model.add(layers.Conv1D(32, 3, activation='relu'))
model.add(layers.MaxPooling1D(2))

# Agregar BatchNormalization después de la capa Conv1D
model.add(layers.BatchNormalization())

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))

# Agregar Dropout antes de la capa Dense
model.add(layers.Dropout(0.5))

model.add(layers.Dense(3, activation='softmax'))

# Usar Adam con una tasa de aprendizaje ajustable
model.compile(optimizer=Adam(lr=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
model.summary()
```

```python
accuracy = model.evaluate(X_test, y_test)
print("Accuracy on test data:", accuracy)
```
