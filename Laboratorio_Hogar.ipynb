### Preparación de Datos

```python
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

#### Hogar

```python
# separación de la data 
```

```python
y_H = y_H.astype(int)
y_H
```

```python
# pasamos la pobreza a categoría
y_H = to_categorical(y_H)
y_H
```

notamos que por algún motivo toma 4 categorías. Eliminamos el primer 0. 

```python
y_H = y_H[:, 1:]
y_H
```

```python
# separación de la data
X_train, X_test, y_train, y_test = train_test_split(X_H, y_H, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# X_train = X_train.values.astype('float32')
# y_train = y_train.astype('float32')
```

### Construcción del modelo

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
model = models.Sequential()
```

```python
model.add(layers.Reshape((127, 1), input_shape=(127,)))
model.add(layers.Conv1D(32, 3, activation='relu'))
model.add(layers.MaxPooling1D(2))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))  
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
model.summary()
```

### Evaluación de Modelo

```python
accuracy = model.evaluate(X_test, y_test)
print("Accuracy on test data:", accuracy)
```
